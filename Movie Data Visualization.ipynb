## Data Visulisation Challenge
Given a Movies Dataset, your task is to make a visualisation that -
- Plots the Length of the Movie Title Names on X-Axis
- Frequency Counts on Y-axis i.e. Number of Movies having 'x' characters in their title


```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read a CSV File

movies_df = pd.read_csv("movie_metadata.csv")
```

```python
movies_df.head(n = 10)
```

```python
movies_df.columns
```

```python
movie_titles = movies_df['movie_title']  # movies_df.get("movie_title")
```

```python
print(movie_titles)
```

```python
print(type(movie_titles))
```

```python
movie_titles = list(movie_titles)
```

```python
movie_titles
```

```python
print(type(movie_titles))
```

```python
print(movie_titles[:5])
```

```python
freq_titles = {}
```

```python
for title in movie_titles:
    len_of_title = len(title)
    
    if freq_titles.get(len_of_title) is None:
        freq_titles[len_of_title] = 1
    else:
        freq_titles[len(title)] += 1 
```

```python
print(freq_titles)
```

```python
X = np.array(list(freq_titles.keys() ) )
Y = np.array(list(freq_titles.values() ) )
```

```python
print(X)
print(Y)
```

```python
plt.scatter(X, Y)
plt.xlabel("Length of Moviel Title")
plt.ylabel("Frequency Count")
plt.title("Movie Data Visulisation Problem")
plt.show()
```
